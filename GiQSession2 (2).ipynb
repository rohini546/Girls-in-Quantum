```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, execute, QuantumRegister,BasicAer
from qiskit.tools.jupyter import *
from qiskit.visualization import plot_state_qsphere, plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator


# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

# Let us create the bell states!


```python
bell_0=QuantumCircuit(2,2)
bell_0.h(0)
bell_0.cx(0,1)
bell_0.draw('mpl')
```

```python
bell_0.measure(0,0)
bell_0.measure(1,1)
backend=Aer.get_backend('qasm_simulator')
counts=(execute(bell_0,backend,shots=1000).result().get_counts())
plot_histogram(counts)
```

# Unitary Representation

```python
bell_0=QuantumCircuit(2,2)
bell_0.h(0)
bell_0.cx(0,1)
bell_0.draw('mpl')
```

```python
back = Aer.get_backend('unitary_simulator')
result = execute(bell_0, back).result()
U = result.get_unitary(decimals=3)
U
```
